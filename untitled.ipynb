## تحليل القيم المفقودة والمتطرفة

### Subtask:
قم بفحص البيانات بحثاً عن القيم المفقودة والمتطرفة وتحديد كيفية التعامل معها.


**Reasoning**:
Calculate missing values and their percentages for each column and print the summary. Then, calculate descriptive statistics for numerical columns to identify potential outliers and print them.



```python
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

missing_summary = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percentage
})

missing_summary = missing_summary[missing_summary['Missing Values'] > 0].sort_values(by='Missing Values', ascending=False)

print("Summary of Missing Values:")
display(missing_summary)

print("\nDescriptive Statistics for Numerical Columns (to identify potential outliers):")
display(df.describe())
```

**Reasoning**:
Summarize the findings from the missing value analysis and the descriptive statistics to determine how to handle missing values and potential outliers in the identified columns.



```python
print("Handling Missing Values Strategy:")
print("- ChurnReason and ChurnCategory have a high percentage of missing values (over 73%). These likely correspond to customers who did not churn, so these missing values can be interpreted as 'No Churn Reason' or 'No Churn Category'. They can be filled with a placeholder like 'Not Applicable' or a similar indicator.")
print("- Offer has over 55% missing values. This likely indicates that no offer was made to these customers. These missing values can also be filled with a placeholder like 'No Offer'.")
print("- InternetType has over 21% missing values. This likely corresponds to customers who do not have internet service. These missing values can be filled with a placeholder like 'No Internet Service'.")

print("\nHandling Potential Outliers Strategy:")
print("Based on the descriptive statistics, most numerical columns like Age, NumberofDependents, Number_of_Referrals, TenureinMonths, AvgMonthlyLongDistanceCharges, AvgMonthlyGBDownload, MonthlyCharge, TotalCharges, TotalRefunds, TotalExtraDataCharges, TotalLongDistanceCharges, TotalRevenue, SatisfactionScore, ChurnScore, and CLTV show a reasonable range between min and max values and between the quartiles. There are no clear indications of extreme outliers that would require immediate removal or transformation without further investigation. Columns like TotalRefunds and TotalExtraDataCharges have a minimum of 0 and a maximum that is significantly higher than the 75th percentile, which is expected as most customers would not have refunds or extra data charges. ZipCode, Latitude, and Longitude represent geographical information and their range is expected. Therefore, for this analysis, we will keep these columns as they are and not perform any outlier removal or transformation based solely on these descriptive statistics. Further investigation into potential outliers might be needed during model building or specific analysis steps if they negatively impact results.")
```

## تحليل العلاقات والارتباطات

### Subtask:
قم بتحليل العلاقات بين الأعمدة المختلفة، بما في ذلك حساب مصفوفة الارتباط للأعمدة الرقمية.


**Reasoning**:
Calculate and display the correlation matrix for numerical columns, then display crosstabs for important categorical columns with 'ChurnLabel'.



```python
numerical_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numerical_df.corr()

print("Correlation Matrix for Numerical Columns:")
display(correlation_matrix)

print("\nCrosstabs for important categorical columns and ChurnLabel:")
important_categorical_cols = ['Gender', 'Contract', 'InternetService', 'PaymentMethod']

for col in important_categorical_cols:
    print(f"\nCrosstab for {col} and ChurnLabel:")
    display(pd.crosstab(df[col], df['ChurnLabel']))

print("\nSummary of key relationships and correlations:")
print("- Numerical Correlation: Examine the correlation matrix for strong positive or negative correlations (e.g., absolute value > 0.5). Note columns with high correlation to 'ChurnScore' or 'CLTV'.")
print("- Categorical Relationships: Examine the crosstabs to see how the distribution of 'Yes' and 'No' in 'ChurnLabel' varies across categories in 'Gender', 'Contract', 'InternetService', and 'PaymentMethod'. Look for categories with a noticeably higher or lower churn rate.")
```

```python
print("Detailed Summary of Key Relationships and Correlations:")

print("\nNumerical Relationships:")
print("- 'TotalCharges', 'TotalLongDistanceCharges', and 'TotalRevenue' are highly positively correlated with 'TenureinMonths', which is expected as customers with longer tenure accumulate higher total charges and revenue.")
print("- 'MonthlyCharge' shows a moderate positive correlation with 'TotalCharges', 'TotalRevenue', and 'TenureinMonths'.")
print("- 'AvgMonthlyGBDownload' shows a negative correlation with 'Age', suggesting younger customers tend to download more data.")
print("- 'ChurnScore' shows a moderate negative correlation with 'TenureinMonths', 'NumberofDependents', and 'SatisfactionScore', indicating that customers who stay longer, have dependents, or are more satisfied are less likely to churn.")
print("- 'CLTV' shows a moderate positive correlation with 'TenureinMonths', 'TotalCharges', and 'TotalRevenue', which is logical as these factors contribute to higher customer lifetime value.")
print("- Geographic information (ZipCode, Latitude, Longitude, Population) shows weak correlations with most other variables, including churn-related metrics.")

print("\nCategorical Relationships with ChurnLabel:")
print("- Gender: The crosstab shows a very similar distribution of churn between Male and Female customers, suggesting Gender is not a strong predictor of churn.")
print("- Contract: Customers with 'Month-to-Month' contracts have a significantly higher churn rate compared to those with 'One Year' or 'Two Year' contracts. 'Two Year' contracts have the lowest churn rate.")
print("- InternetService: Customers with Internet Service ('Yes') have a much higher churn rate than those without Internet Service ('No').")
print("- PaymentMethod: Customers using 'Bank Withdrawal' as a payment method appear to have a higher churn rate compared to those using 'Credit Card' or 'Mailed Check'.")

print("\nOverall Insights on Relationships:")
print("Tenure, Contract type, Internet Service, and Payment Method appear to be significant factors related to customer churn. Longer tenure, longer contract terms, no internet service, and certain payment methods are associated with lower churn rates. Numerical metrics like TotalCharges, TotalRevenue, and CLTV are strongly linked to tenure, as expected.")
```

## اقتراح التصورات

### Subtask:
اقترح أكواد Python لإنشاء تصورات مناسبة لتوضيح أهم النتائج والرؤى.


**Reasoning**:
Generate Python code to create visualizations for the relationships between Contract and ChurnLabel, InternetService and ChurnLabel, TenureinMonths and TotalRevenue, SatisfactionScore and ChurnLabel, and a heatmap of the correlation matrix for key numerical columns.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a larger figure size for better readability of plots
plt.figure(figsize=(15, 10))

# 1. Relationship between Contract and ChurnLabel
plt.subplot(2, 3, 1) # 2 rows, 3 columns, 1st plot
sns.countplot(data=df, x='Contract', hue='ChurnLabel', palette='viridis')
plt.title('Churn Rate by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Number of Customers')

# 2. Relationship between InternetService and ChurnLabel
plt.subplot(2, 3, 2) # 2 rows, 3 columns, 2nd plot
sns.countplot(data=df, x='InternetService', hue='ChurnLabel', palette='viridis')
plt.title('Churn Rate by Internet Service')
plt.xlabel('Internet Service')
plt.ylabel('Number of Customers')

# 3. Relationship between TenureinMonths and TotalRevenue
# Using a scatter plot to show the relationship
plt.subplot(2, 3, 3) # 2 rows, 3 columns, 3rd plot
sns.scatterplot(data=df, x='TenureinMonths', y='TotalRevenue', alpha=0.6, color='skyblue')
plt.title('Tenure vs. Total Revenue')
plt.xlabel('Tenure in Months')
plt.ylabel('Total Revenue')

# 4. Distribution of SatisfactionScore by ChurnLabel
plt.subplot(2, 3, 4) # 2 rows, 3 columns, 4th plot
sns.boxplot(data=df, x='ChurnLabel', y='SatisfactionScore', palette='viridis')
plt.title('Satisfaction Score Distribution by Churn')
plt.xlabel('Churn Label')
plt.ylabel('Satisfaction Score')

# 5. Heatmap of Correlation Matrix for key numerical columns
plt.subplot(2, 3, (5, 6)) # Span across 5th and 6th positions
# Selecting a subset of relevant numerical columns for the heatmap
# Including ChurnScore and CLTV as requested
key_numerical_cols = ['TenureinMonths', 'MonthlyCharge', 'TotalCharges', 'TotalRevenue',
                      'SatisfactionScore', 'ChurnScore', 'CLTV', 'NumberofDependents',
                      'Number_of_Referrals', 'AvgMonthlyGBDownload', 'AvgMonthlyLongDistanceCharges',
                      'TotalRefunds', 'TotalExtraDataCharges', 'Age', 'Population']
correlation_subset = df[key_numerical_cols].corr()
sns.heatmap(correlation_subset, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Key Numerical Features')
plt.tight_layout() # Adjust layout to prevent overlapping
plt.show()
```
